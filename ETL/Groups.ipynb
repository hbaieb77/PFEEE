```python
import re 
from pyspark.sql import SparkSession
from pyspark.sql.types import StringType, ArrayType, StructType, StructField
from pyspark.sql.functions import regexp_replace, col, udf, explode




df_group_1_1 = df_group_1_1.withColumn('Employee du groupe', regexp_replace(col('Employee du groupe'), r'[^\w\s,]', ''))
df_group_1_1 = df_group_1_1.withColumn('Employee du groupe', regexp_replace(col('Employee du groupe'), r'\s+', ''))
df_group_1_1 = df_group_1_1.withColumn('Employee du groupe', regexp_replace(col('Employee du groupe'), r'LoginNom,Prnom', ''))
df_group_1_1 = df_group_1_1.withColumn('Employee du groupe', regexp_replace(col('Employee du groupe'), r'\s+', ''))

# UDF to extract login IDs and names
def extract_login_and_names(value):
    if isinstance(value, str):
        login_id_parts = re.findall(r'(M\d{5}|MA\d{4})', value)
        login_ids_row = []
        nom_prenoms_row = []
        for login_id in login_id_parts:
            match = re.search(rf'{login_id}([a-zA-Z]+),([a-zA-Z]+)', value)
            if match:
                login_ids_row.append(login_id)
                nom_prenom = match.group(1) + ' ' + match.group(2)
                if nom_prenom.endswith('M'):
                    nom_prenom = nom_prenom[:-1]
                elif nom_prenom.endswith('MA'):
                    nom_prenom = nom_prenom[:-2]
                nom_prenoms_row.append(nom_prenom)
        return (login_ids_row, nom_prenoms_row)
    return ([], [])

extract_login_and_names_udf = udf(extract_login_and_names, StructType([
    StructField("Login IDs", ArrayType(StringType()), True),
    StructField("Nom, Prenoms", ArrayType(StringType()), True)
]))

# Apply UDF and split into separate columns
df_group_1_1 = df_group_1_1.withColumn("LoginNames", extract_login_and_names_udf(col("Employee du groupe")))
df_group_1_1 = df_group_1_1.withColumn("Login IDs", col("LoginNames").getItem("Login IDs"))
df_group_1_1 = df_group_1_1.withColumn("Nom, Prenoms", col("LoginNames").getItem("Nom, Prenoms"))
df_group_1_1 = df_group_1_1.drop("LoginNames")

# Explode Login IDs and Nom, Prenoms for creating employee and employee_group DataFrames
df_exploded = df_group_1_1.withColumn("Login ID", explode(col("Login IDs")))\
                          .withColumn("Nom, Prenom", explode(col("Nom, Prenoms")))

# Create employees DataFrame
df_employees = df_exploded.select(col("Login ID").alias("ID"), col("Nom, Prenom").alias("Name_LastName")).distinct()

# Create employee_group DataFrame
df_employee_group = df_exploded.select(col("Login ID").alias("Employee_ID"), col("GROUP_ID").alias("Group_ID")).distinct()
```
